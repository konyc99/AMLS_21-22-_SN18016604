```python
labels = pd.read_csv("dataset/image/label_binary2.csv", index_col=0)

labels.head()

def get_img(filename, directory="dataset/image/"):
    
    file_path = os.path.join(directory, filename)
    img = Image.open(file_path)
    return np.array(img)


no_tumor_row = labels[labels.label == 0.0].index[67]

plt.imshow(get_img(no_tumor_row))
plt.show()

tumor_row = labels[labels.label == 1.0].index[67]

plt.imshow(get_img(tumor_row))
plt.show()

print(tumor_row)
```

```python
# load a bombus image using our get_image function and bombus_row from the previous cell
tumor = get_img(tumor_row)

print('RGB tumor scan image has shape: ', tumor.shape)


grey_tumor = rgb2grey(tumor)

plt.imshow(grey_tumor, cmap=mpl.cm.gray)

print('Greyscale tumor image has shape: ', grey_tumor)

hog_features, hog_image = hog(grey_tumor,
                              visualize=True,
                              block_norm='L2-Hys',
                              pixels_per_cell=(16, 16))

plt.imshow(hog_image, cmap=mpl.cm.gray)
```

```python
def create_features(img):
    
    #resize image for data efficient analysis
    
    resized_img = resize(img, (128, 128))
    
    # flatten three channel color image
    color_features = img.flatten()
    # convert image to greyscale
    grey_image = rgb2grey(img)
    # get HOG features from greyscale image
    hog_features = hog(grey_image, block_norm='L2-Hys', pixels_per_cell=(16, 16))
    # combine color and hog features into a single array
    flat_features = np.hstack(color_features)
    return flat_features

tumor_features = create_features(resized_img)

print(tumor_features.shape)
```

```python
def create_feature_matrix(label_dataframe):
    features_list = []
    
    for img_id in label_dataframe.index:
        # load image
        img = get_img(img_id)
        # get features for image
        image_features = create_features(img)
        features_list.append(image_features)
        
    # convert list of arrays into a matrix
    feature_matrix = np.array(features_list)
    return feature_matrix

```

```python
# run create_feature_matrix on our dataframe of images
feature_matrix = create_feature_matrix(labels)
```
