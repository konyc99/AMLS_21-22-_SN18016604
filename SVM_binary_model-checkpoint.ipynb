```python
import os

import matplotlib as mpl
import matplotlib.pyplot as plt


import pandas as pd
import numpy as np

from PIL import Image

from skimage.feature import hog
from skimage.color import rgb2gray

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from skimage.transform import resize

from sklearn.metrics import roc_curve, auc
import plot_learning_curve as plots
```

```python
#import labels from the csv file
#labels have been changed to boolean 1 and 0 for simplicity
labels = pd.read_csv("dataset/image/label_binary2.csv", index_col=0)

labels.head()

#function responsible for fetching the image from the dataset and turning
#into an array
def get_img(filename, directory="dataset/image/"):
    
    file_path = os.path.join(directory, filename)
    img = Image.open(file_path)
    return np.array(img)

#test of label indexing by tumor = 1  and no_tumor = 0
no_tumor1 = labels[labels.label == 0].index[67]

#show an image of no tumor - display testing capabilities
plt.imshow(get_img(no_tumor1))
plt.show()

#same as above but for tumor example
tumor1 = labels[labels.label == 1].index[67]

plt.imshow(get_img(tumor1))
plt.show()
```

```python
#fetch the image displayed above
tumor_img = get_img(tumor1)

print('The tumor image as an array has a shape of: ', tumor_img.shape)
#As seen above despite looking greyscale the CAT scans are RGB images
#with 3 channeles that need to be flattened

#simple greyscale transformation
grey_tumor = rgb2gray(tumor_img)

print('The tumor image in greyscale has a shape of: ', grey_tumor.shape)
#Flattened image will have a shape of (512,512,1) as the RGB channeles
#chage to just a single greyscale channel


#This is a show of features extraction just to understand what does it
#actually mean and how is it relevant to what we're doing
#HOG - Histogram Oriented Gradients
hog_features, hog_image = hog(grey_tumor,
                              pixels_per_cell=(16, 16),
                              #normalizaton using L2 norm and limiting the max values to 0.2
                              #(Hys - hysteresis)
                              block_norm='L2-Hys',
                              #returns image to be able to plot and see it
                              visualize=True
                              )
#show HOG image visualisation
plt.figure(figsize = (10,10))
#plt.imshow(hog_image, cmap=mpl.cm.magma)
plt.imshow(hog_image, cmap=mpl.cm.gray)
```

```python
#this function takes image and target size if specified as input
#and outputs image features extracted using HOG described earlier
def create_features(img,image_size=None):
    
    #resize image into the target size if size specified
    if type(image_size) == tuple:
        img = resize(img, (image_size))    
    
    # convert image to greyscale for the HOG algorirthm
    grey_image = rgb2gray(img)
    
    # acquire HOG features using the greyscaled imgage
    hog_features = hog(grey_image, block_norm='L2-Hys', pixels_per_cell=(16, 16))
    
    return hog_features

#test feature extraction
tumor_features = create_features(tumor_img)
print(tumor_features.shape)
```

```python
#tqdm library is for progress tracking of the feature matrix creation
from tqdm import tqdm

def create_feature_matrix(label_dataframe,target_size=None):
    features_list = []
    
    #iteration through label dataframe to pull all images
    for img_id in tqdm(label_dataframe.index):
        # image loading given label
        img = get_img(img_id)
        # create features for the given image using the predefined function
        image_features = create_features(img,target_size)
        #append the existing list as of features
        features_list.append(image_features)
        
    #convert the list into a np array in the given shape
    feature_matrix = np.array(features_list)
    return feature_matrix

```

```python
#testing and visualising hown different resize operations are affecting image quality of the scans
#Theory there's something to simplifing the image for better feature extraction

fig, axs = plt.subplots(2,2,figsize = (10,10))
axs[0, 0].imshow(tumor_img)
axs[0, 0].set_title('512x512')

tumor256 = resize(tumor_img, (256, 256))
axs[1, 0].imshow(tumor256)
axs[1, 0].set_title('256x256')

tumor128 = resize(tumor_img, (128, 128))
axs[0, 1].imshow(tumor128)
axs[0, 1].set_title('128x128')

tumor64 = resize(tumor_img, (64, 64))
axs[1, 1].imshow(tumor64)
axs[1, 1].set_title('64x64')
```

```python
# create feature matrix given labels and output shape of the data
feature_matrix = create_feature_matrix(labels)
print(feature_matrix.shape)
```

```python
def PCA_transform(features):  
    # Standardize features by removing the mean and scaling to unit variance.
    # Define a standard  scaler
    sscal = StandardScaler()

    # run the scalar on the feature matrix
    feature_stand = sscal.fit_transform(features)

    # Principal component analysis - Linear dimensionality reduction using Singular Value Decomposition
    # The input data is centered but not scaled - that's why we apply SS first
    pca = PCA(n_components = 0.95)
    pca.fit(feature_stand)


    # Apply PCA on the feature matrix using fir_transform function
    feature_pca = pca.transform(feature_stand)

    #compare the shapes after the feature preprocessig
    print('Feature matrix before preprocessing: ', features.shape)
    print('Feature matrix after PCA transform is: ', feature_pca.shape)
    return feature_pca

new_feature_matrix = PCA_transform(feature_matrix)
```

```python
#Split the feature data into x/y and train/test using the scikit train_test_split
x= pd.DataFrame(new_feature_matrix)
y=pd.Series(labels.label.values)

X_train, X_test,Y_train,Y_test = train_test_split(x,
                                                  y,
                                                  test_size=.2,
                                                  random_state=67)
```

```python
#train a linear model using SVC
svm = SVC(kernel = 'linear', probability =True, random_state = 67)

svm.fit(X_train, Y_train)
```

```python
#train a sigmoid model using SVC
svm2 = SVC(kernel = 'rbf', probability =True, random_state = 67)

svm2.fit(X_train, Y_train)
```

```python
#use predict function to get algorithms guesses on th e test set
Y_pred = svm.predict(X_test)
Y_pred2 = svm2.predict(X_test)

#Run an accuracy test for both models
acc = accuracy_score(Y_test, Y_pred)
acc2 =accuracy_score(Y_test, Y_pred2)

print('Given model (linear) achieves accuracy of: ', acc)
print('Given model (RBF) achieves accuracy of: ', acc2)
```

```python
plots.plot_learning_curve(estimator = svm,title = 'Learning Curve linear',X=X_train,y=Y_train)
plots.plot_learning_curve(estimator = svm2,title = 'Learning Curve RBF',X=X_train,y=Y_train)
```
